```python
url ="https://www.linkedin.com/jobs/search/?currentJobId=3857355048&f_I=96&f_TPR=r86400&f_WT=1%2C3%2C2&geoId=91000000&location=European%20Union&origin=JOB_SEARCH_PAGE_JOB_FILTER&refresh=true&sortBy=R"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
# soup
```

```python
# location = soup.find_all('span', class_='job-search-card__location')
# location
```

```python
def get_location(soup):
    location_elements = soup.find_all('span', class_='job-search-card__location')
    locations = [location.text.strip() for location in location_elements]
    return locations
# get_location(soup)
```

```python
def get_title(soup):
    title_elements = soup.find_all('h3', class_='base-search-card__title')
    # Extract text from each title element
    titles = [title.text.strip() for title in title_elements]
    return titles
# get_title(soup)
```

```python
def get_company(soup):
    company_elements = soup.find_all('h4', class_='base-search-card__subtitle')
    companies = [companies.text.strip() for companies in company_elements]
    return companies
# get_company(soup)
```

```python
def extract_job(soup,jobtype):
    titles = get_title(soup)
    companies = get_company(soup)
    locations = get_location(soup)
    df = pd.DataFrame({
        'title': titles,
        'company': companies,
        'location': locations      
    })
    column_name = jobtype
    df[column_name]= jobtype
    return df

mix_scrape = extract_job(soup,'M')
mix_scrape
```

```python
merged_df1 = mix_scrape.merge(remote_scrape, on='title', how='left')
merged_df1.drop(columns=['company_y', 'location_y'], inplace=True)
merged_df1
```

```python
merged_df2 = merged_df1.merge(hybrid_scrape, on='title', how='left')
merged_df2.drop(columns=['company', 'location'], inplace=True)
merged_df2
```

```python
pd.set_option('display.max_rows', None)

merged_df3 = merged_df2.merge(onsite_scrape, on='title', how='left')
merged_df3.drop(columns=['company', 'location'], inplace=True)
merged_df3
```

```python
merged_df3.dropna(subset=['R', 'H', 'S'], how='all', inplace=True)
merged_df3
```
